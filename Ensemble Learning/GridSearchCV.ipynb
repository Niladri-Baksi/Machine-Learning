```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species',axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier() #for k=13 the model is overfitting
```

```python
model_knn.fit(X_train,y_train)
```

```python
# y_pred = model_knn.predict(X_test)
# accuracy_score(y_test,y_pred)
```

```python
model_knn.score(X_test,y_test) #gives the same thing as above two lines in the previous cell. It is a shortcut for the above two lines. It gives the accuracy score directly.
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma= 'auto') #overfitting for C=10
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

# GridSearchCV

```python
#now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier_SVM = GridSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},cv = 5,return_train_score = True)
```

```python
classifier_SVM.fit(X,y)
```

```python
classifier_SVM.cv_results_
```

```python
results_SVM = pd.DataFrame(classifier_SVM.cv_results_)
```

```python
results_SVM
```

```python
results_SVM.columns
```

```python
results_SVM[['param_C','param_kernel','mean_test_score','mean_train_score']] #mean train vs mean test scores gives shows if the model is overfitting or not. If the mean train score is much higher than the mean test score then the model is overfitting. If they are close to each other then the model is not overfitting.
```

```python
classifier_SVM.best_params_ #gives the best parameters for the model. It gives the parameters that gives the highest mean test score.
```

```python
classifier_knn = GridSearchCV((model_knn),{
    'n_neighbors' : [1,3,5,7,9,11,13],
    'weights' : ['uniform','distance'],
    'algorithm' : ['auto','ball_tree','kd_tree','brute'],
    'metric' : ['euclidean','manhattan','minkowski']
},cv = 7,return_train_score = True)
```

```python
classifier_knn.fit(X,y)
```

```python
results_knn = pd.DataFrame(classifier_knn.cv_results_)
```

```python
results_knn
```

```python
results_knn[['param_n_neighbors','mean_test_score','mean_train_score']]
```

```python
classifier_knn.best_params_
```
