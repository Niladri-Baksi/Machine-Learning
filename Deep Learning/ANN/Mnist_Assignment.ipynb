```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255 #rescaling each pixels
```

```python
X_train_flatten = X_train.reshape(len(X_train),28*28)
X_train_flatten.shape #flatten to 1D
```

```python
X_test_flatten = X_test.reshape(len(X_test),28*28)
X_test_flatten.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, input_shape=(784,), activation='relu'), # Hidden layer 1
    #keras.layers.Dense(128, activation='relu'), 
    keras.layers.Dense(64, activation='relu'), # Hidden layer 2
    keras.layers.Dense(32, activation='relu'), # Hidden layer 3
    keras.layers.Dense(10, activation='softmax') # Output layer
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history=model.fit(X_train_flatten, y_train, epochs=15,batch_size=32) #batch size 32 is the default value, it means that the model will update its weights after processing 32 samples. This can help to speed up training and can also help to prevent overfitting. However, the optimal batch size can vary depending on the specific dataset and model architecture, so it may be worth experimenting with different values to see which one works best for your particular case.
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
pred=model.predict(X_test_flatten)
```

```python
plt.imshow(X_test[9999], cmap='gray')
plt.title(f"Predicted: {pred[9999].argmax()}, Actual: {y_test[9999]}")
plt.axis('off')
plt.show()
```

```python
for i in range(10):
    plt.imshow(X_test[i], cmap='gray')
    plt.title(f"Predicted: {pred[i].argmax()}, Actual: {y_test[i]}")
    plt.axis('off')
    plt.show()
```

```python
# Plot accuracy
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
#plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
#plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
y_predicted= model.predict(X_test_flatten)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sns.heatmap(cm,annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# Readme

#### What your data actually looks like
After flattening:

X_train_flatten.shape

#### Output:

(60000, 784)
What this means:
60,000 images
Each image = 784 features

#### Think of it like a table:

Image1 → [784 values]
Image2 → [784 values]
Image3 → [784 values]
...
Image60000 → [784 values]

#### How Keras sees this
When you do:
model.fit(X_train_flatten, y_train, epochs=15)
You are passing:
X_train_flatten → ALL 60,000 images
y_train → ALL labels

#### What happens internally (VERY IMPORTANT)
Keras does NOT train on everything at once blindly.

It uses something called batch processing
🔹 Default behavior
If you don’t specify batch size:

model.fit(...)

#### Default:

batch_size = 32
🔁 Training flow
Total images = 60000
Batch size = 32

#### So:

60000 / 32 ≈ 1875 batches per epoch
For EACH batch:
Take 32 images
   ↓
Each image = 784 input
   ↓
Pass through network
   ↓
Compute loss
   ↓
Update weights
🧠 So what goes into the model?
Not this ❌
[784]   (single image)
But this ✅
(32, 784)   ← one batch

So, Model processes multiple images at once

#### Inside the model (VERY IMPORTANT)

The first layer:

Dense(128, input_shape=(784,))

This means:

Each input sample = 784 features
But model can handle many samples together
Internally it becomes:
Input batch: (32, 784)
Weights:     (784, 128)

Matrix multiplication:

(32, 784) × (784, 128) → (32, 128)

That’s how ALL images are processed efficiently

#### Simple Analogy

Think of:

Each image = one student answer sheet
Batch = 32 answer sheets

Teacher (model) checks:

32 sheets at once
Learns from mistakes
Updates marking scheme (weights)
#### Full Training Flow
Epoch 1:
   Batch 1 (32 images)
   Batch 2 (32 images)
   ...
   Batch 1875

Epoch 2:
   Repeat again
#### Important Clarification
Flattening does NOT reduce dataset size

It only changes:

(28, 28) → (784)

You still have:

60000 images
🚀 Visual Summary
Before flatten:
(60000, 28, 28)

After flatten:
(60000, 784)

Training:
→ Sent in batches like (32, 784)
→ Not a single image
#### Bonus (VERY IMPORTANT for interviews)

If they ask:

“How does Keras handle multiple samples?”

You say:

“Using batch processing with vectorized operations (matrix multiplication), which allows efficient parallel computation.”
