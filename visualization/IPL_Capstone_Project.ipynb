### Loading the Libraries and Dataset

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings("ignore")

df = pd.read_csv( 'IPL.csv')
df

#df.head()

```

Now let's see how many columns have null values in total.

```python
df.isnull()
df.isnull().sum() #to get the count of null values in each column...if false then there are no null values in that column else there are null values in that column and the count of those null values is given by the sum function since 0+1=1 and 0+0=0
```

**Now, Here comes some Basic Questions**

1. Which team won the most matches?

```python
match_wins=df['match_winner'].value_counts()
match_wins
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')#since match wins is a series and its index is the team name and its values are the count of match wins for each team so we are plotting index on y axis and values on x axis
plt.title("Most match won by team ")
plt.xlabel("Number of matches won")
plt.ylabel("Teams")
```

2. Toss Decision Trends

```python
sns.countplot(x = df['toss_decision'],palette='rainbow')
plt.title("Toss Decision Trends ")
```

3. Toss Winner vs Match Winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count() #to get the count of matches where toss winner and match winner are same we are using boolean indexing to filter the dataframe and then counting the number of match ids in that filtered dataframe...match id is a unique identifier for each match so counting the number of match ids gives us the count of matches where toss winner and match winner are same instead of getting the count in all the columns
percentage = (count *100)/df.shape[0]
percentage.round(2)
```

4. How do teams win? (Runs vs Wickets)

```python
sns.countplot(x = df['won_by'], palette='Set2')
plt.title("Won by")
```

###  Key Player Performances

1 Most "Player of the Match" Awards

```python
df.head()
```

```python
count = df['player_of_the_match'].value_counts().head(10) 
count
```

```python
sns.barplot(x = count.values , y = count.index,palette='mako')
plt.title("Top 10 players with man of the match")
```

2 Top Scorers

```python
df.head()
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2) #to get the top scorers and their top runs we groupby and aggregate(sum) all their runs of matches in which they were the top scorers and then sort those runs in descending order
high
```

```python
high.plot(kind = 'barh', color = 'orange') #barh is used to plot horizontal bar graph 
```

10 Best Bowling Figures

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'bar', color = 'green')
```
